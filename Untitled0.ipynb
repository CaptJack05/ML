```python
non_smoker_df = medical_df[medical_df.smoker == "no"]

non_smoker_df
```

```python
plt.title("age vs charges")
sns.scatterplot(data = non_smoker_df, x = 'age', y= 'charges',alpha = 0.7, s = 15)
fig = px.scatter(
    non_smoker_df,
    x = 'age',
    y = 'charges',
)
fig.show()
```

```python
def estimateCharge(age,w , b):
  return w * age + b
```

```python
w = 300
b = 200

print(estimateCharge(30, w, b))
```

```python
ages = non_smoker_df.age
estimated_charges = estimateCharge(ages , w, b)
estimated_charges
```

```python
plt.plot(ages,estimated_charges,'r');
plt.xlabel('Ages');
plt.ylabel('Estimated Charges')
```

Plotting Modedel and dta together

```python
def tryParam(w,b):
  ages = non_smoker_df.age
  charges = non_smoker_df.charges
  estimated_charges = estimateCharge(ages, w, b)

  plt.scatter(ages, charges, s=8);
  plt.plot(ages,estimated_charges,'r', alpha = 0.8);
  plt.xlabel("ages");
  plt.ylabel("Charges");
  plt.legend(['Estimated','Actual'])
  loss = rmse(target, estimated_charges)
  print("RMSE Loss: ", loss)
```

```python
target = non_smoker_df.charges

plt.plot(ages, estimated_charges, 'r', alpha = 0.9)
plt.scatter(ages, target, s = 8, alpha = 0.8)
plt.xlabel("Ages")
plt.ylabel("Charges")
plt.legend(['Estimate','Actual'])
```

```python
tryParam(300,200)
```

```python
tryParam(500,-2000)
```

```python
import numpy as np
def rmse(target,prediction):  # target = actual data  prediction = model output
  return(np.sqrt(np.mean(np.square(target - prediction))))
```

```python
target = non_smoker_df['charges']
predicted = estimateCharge(non_smoker_df.age , w, b)

rmse(target,predicted)
```
